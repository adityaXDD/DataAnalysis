### TASK 3 : Price Range vs. Online Delivery and Table Booking

### Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python

cross_tab = pd.crosstab(index=df['Price range'], 
                        columns=[df['Has Online delivery'], df['Has Table booking']], margins=True)

cross_tab


```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cross_tab,cmap='coolwarm',fmt = 'd',annot=True)
plt.title('Relationship b/w Price range to Online Delivery and Table Booking')
plt.xlabel('Has Online Delivery and Table Booking')
plt.show()
```

### Determine if higher-priced restaurants are more likely to offer these services.

```python
grouped_data = df.groupby('Price range').agg({
    'Has Online delivery': lambda x: (x == 'Yes').mean() * 100,
    'Has Table booking': lambda x: (x == 'Yes').mean() * 100
})

grouped_data
```

```python
grouped_data.plot( kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Percentage of Restaurants offering Service')
plt.xlabel('Price Range')
plt.ylabel('Percentage')

plt.show()
```
