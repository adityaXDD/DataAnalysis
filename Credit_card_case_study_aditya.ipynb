### Merging the tables

```python
csr = pd.merge(left=customer_acquisition,right=pd.merge(left=spend,right=repayment,
    left_on=['Sl No:','Customer'],right_on=['SL No:','Customer']),left_on='Customer',right_on='Customer')
```

```python
csr.head()
```

```python
csr.drop(columns=['Sl No:','SL No:'], inplace=True)
```

#### b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card)

```python
csr.Spending_Amount = csr.Spending_Amount.where(csr.Spending_Amount < csr.Limit, csr.Limit/2)
```

#### c. Incase the repayment amount is more than the limit, replace the repayment with the limit.


```python
csr.Repayment_Amount = csr.Repayment_Amount.where(csr.Repayment_Amount < csr.Limit ,csr.Limit)
```

### 5. Write user defined PYTHON function to perform the following analysis: You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.

```python
csr
```

```python
csr['repay_year']= csr.Repayment_Date.dt.year
```

```python
def top10cust(prod, timeframe):
    """
    Retrieve the top customers for each city for a given product within a specified timeframe.

    Parameters:
    - prod (str): The product for which to retrieve top customers.['Gold/Silver/Platinum']
    - timeframe (str): The time period for which to analyze customer data.['Monthly'/'Yearly']

    Returns:
    list: A list of the top customers for the specified product and timeframe for each city.

    """
    pd.set_option('display.max_rows',500)
    
    tempdf = csr[csr.Product == prod]
    if timeframe == 'Monthly':
        res = tempdf.groupby(['repay_month','City','Customer'])['Repayment_Amount'].sum()
    elif timeframe == 'Yearly':
        res = tempdf.groupby(['repay_year','City','Customer'])['Repayment_Amount'].sum()
    return res
```

```python
top10cust('Gold','Yearly')
```
